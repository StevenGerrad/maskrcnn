## 删除某些参数文件
> keras的fit_generator函数中的keras.callbacks.ModelCheckpoint默认每个epoch保存参数，占用空间

```python
import os

LOGS_DIR = './logs/faces20210124T1057/'

weights_list = os.listdir(LOGS_DIR)
# print(weights_list)
```

```python
import re

regex = r"mask\_rcnn\_[\w-]+(\d{4})\.h5"
for i in weights_list:
    m = re.match(regex, i)
    if m:
        epoch_num = int(m.group(1))
        # 保留100整数倍的权值文件
        if epoch_num % 100 != 0:
            os.remove(os.path.join(LOGS_DIR, i))
```
